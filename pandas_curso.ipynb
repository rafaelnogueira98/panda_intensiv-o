```python
def low_values_red(value):
    color = "red" if value < 92 else "black"
    return "color: %s" %color # "color: red" ou "color: black"

fifa19[["Overall", "Potential"]].head(10).style.applymap(low_values_red)
```

```python
fifa19.head(10).style.applymap(low_values_red, subset=["Overall", "Potential"])
```

```python
def highlights_max(s):
    is_max = s == s.max() # Checa o booleano s == s_max(), e depois guardar em is_max
    return ["background-color: yellow" if v else "" for v in is_max]

fifa19.head(10).style.apply(highlights_max, subset=["Age", "Overall", "Potential"])
```

```python
fifa19[["Overall", "Potential"]].head(10).style.\
applymap(low_values_red).\
apply(highlights_max)
```

```python
fifa19["Weight"].head()
```

```python
fifa19["Weight"].fillna("0lbs", inplace=True)
```

```python
def substitui_lbs(value):
    """Função para tratar campo peso"""
    x = value.replace("lbs", "")
    return int(x)

fifa19["New Weight"] = fifa19["Weight"].apply(substitui_lbs)
fifa19.head()
```

```python
def substitui_E_M(value):
    x = value.replace("€", "").replace(".", "").replace("M", "00000").replace("K", "000")
    return int(x)

fifa19_no_missing = fifa19["Release Clause"].dropna()
fifa19["New Release Clause"] = fifa19_no_missing.astype(str).apply(substitui_E_M)
fifa19.head()
```

```python
fifa19["avg_GK"] = (fifa19["GKDiving"]+fifa19["GKHandling"]+fifa19["GKKicking"]+fifa19["GKPositioning"]+fifa19["GKReflexes"])/5
fifa19.head()
```

```python
# Criar um campo condicional a outro
fifa19["age_class"] = np.where(fifa19["Age"]>30, "Old", "Young") # import numpy as np
fifa19.head()
# Outra maneira de fazer isso seria:
# fifa19["age_class"] = "Young"
# fifa19.loc(fifa19["Age"]>30, "age_class") = "Old"
```
